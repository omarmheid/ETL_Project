```python
import pandas as pd 
import numpy as np
from sqlalchemy import create_engine 
```

```python
imports = "Resources/Jordan_Import.csv"
imports_df = pd.read_csv(imports)
imports_df.head()
```

```python
imports_transformed = imports_df[["Reporter Name", "Partner Name", "Year","Trade Flow","Product Group","Import (US$ Thousand)","Import Product Share (%)"]]
imports_transformed = imports_transformed.rename(columns={"Reporter Name": "Country",
                                                         "Partner Name": "Partner_Country",
                                                         "Year": "Year",
                                                         "Product Group": "Product_Group",
                                                         "Import (US$ Thousand)": "Import_Thousands($)",
                                                         "Import Product Share (%)": "Import_Share(%)"})
imports_transformed.set_index("Country", inplace=True)
imports_transformed
```

```python
imports_transformed = imports_transformed.sort_values(by = "Import_Thousands($)", ascending = False)
imports_transformed
```

```python
exports = "Resources/Jordan_Export.csv"
exports_df = pd.read_csv(exports)
exports_df
```

```python
exports_transformed = exports_df[["Reporter Name", "Partner Name", "Year","Trade Flow","Product Group","Export (US$ Thousand)","Export Product Share (%)"]]
exports_transformed = exports_transformed.rename(columns={"Reporter Name": "Country",
                                                         "Partner Name": "Partner_Country",
                                                         "Year": "Year",
                                                         "Product Group": "Product_Group",
                                                         "Export (US$ Thousand)": "Export_Thousands($)",
                                                         "Export Product Share (%)": "Export_Share(%)"})
exports_transformed.set_index("Country", inplace=True)
exports_transformed
```

```python
exports_transformed = exports_transformed.sort_values(by = "Export_Thousands($)", ascending = False)
exports_transformed
```

```python
jordan_export_import = pd.merge(left = imports_transformed, right = exports_transformed, left_on = "Country", right_on = "Country")
jordan_export_import = jordan_export_import.rename(columns={"Country": "Country",
                                                          "Partner_Country_x": "Partner_Country_Import",
                                                          "Year_x": "Year_Import",
                                                          "Trade_Flow_x": "Trade_Flow_Import",
                                                          "Product_Group_x": "Product_Group_Import",
                                                          "Import_Thousands($)": "Import_Thousands",
                                                          "Import_Share(%)": "Import_Share",
                                                          "Partner_Country_y": "Partner_Country_Export",
                                                          "Year_y": "Year_Export",
                                                          "Trade_Flow_y": "Trade_Flow_Export",
                                                          "Product_Group_y": "Product_Group_Export",
                                                          "Export_Thousands($)": "Export_Thousands",
                                                          "Export_Share(%)": "Export_Share"})
jordan_export_import
```
